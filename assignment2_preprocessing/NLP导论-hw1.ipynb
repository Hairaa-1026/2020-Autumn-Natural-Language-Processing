Step 1 Tokenization:

```python
import nltk.tokenize as tk
import json


file = open("Gift_Cards_5.json", 'r', encoding='utf-8')
#   原始文本
docs = []
#   由于文件中有多行，直接读取会出现错误，因此一行一行读取
for line in file.readlines():
    dic = json.loads(line)
    for comment in dic:
        if 'reviewText' in comment:
            docs.append(dic[comment])

sentences = []
TokenedComments = []

#   分句：将每条评论按句分割，每个doc代表每条评论的全部文本
for doc in docs:
    sentences.append(tk.sent_tokenize(doc))

#   分词：将每条评论中的每句按词分词，为清晰表示打印出了评论编号及评论内部的句编号
i = 0
for tokens in sentences:
    print('Comment', i, end=':\n')
    i += 1
    j = 0
    sentence = []
    for token in tokens:
        print('Sentence', j, end=':')
        j += 1
        tokens = tk.word_tokenize(token)
        sentence.append(tokens)
        print(tokens)
    TokenedComments.append(sentence)
```

Step 2 Normalization:

```python
#   标准化：消除标点符号、大小写转换、去停用词等
import re
from nltk.corpus import stopwords

NormalizedComments = []
i = 0
for tokens in TokenedComments:
    print('Comment', i, end=':\n')
    i += 1
    j = 0
    sentences = []
    for token in tokens:
        print('Sentence', j, end=':')
        j += 1
        sentence = []
        for word in token:
            pattern = re.compile("[^a-zA-Z0-9\n ]")
            word = re.sub(pattern, "", word).lower()
            if word != '' and word not in stopwords.words('english'):
                sentence.append(word)
        print(sentence)
        sentences.append(sentence)
    NormalizedComments.append(sentences)
```
